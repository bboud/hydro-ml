```python
import sys
sys.path.append('../hydroml')

from dataset import EnergyDensityDataset
from utils import trim

import matplotlib.pyplot as plt

import numpy as np
```

### Check for high values

```python
def plot_anomalies(dataset, threshhold=150):

    for i, data in enumerate(dataset):
        #Trim down the data to the section that we want.
        trim_final_axis, trim_final = trim(dataset.final_eta, data[1], -4.9, -3.1)

        integrated_final = np.trapz( trim_final, trim_final_axis )

        if integrated_final > threshhold:
            fig = plt.figure()
            plt.title(f'Event {i}')
            plt.plot(dataset.final_eta, data[1], color='red')
            plt.plot(dataset.start_eta, data[0], color='blue')
            plt.show()
```

### Data September 5th, 2022

```python
dataset = EnergyDensityDataset('../data/dE_data_3/dE_detas_initial', '../data/dE_data_3/dET_deta_final').smooth().cosh().trim(-6.8,8)
```

```python
plot_anomalies(dataset)
```
